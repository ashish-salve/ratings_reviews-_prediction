### Flipkart.com 

```python
def flipkart_srch(device_type):
    """ This function searches entered device types on Flipkart.com """
    open_url("https://www.flipkart.com")
    #close the login window
    driver.find_element_by_xpath("//div[@class='_2QfC02']/button").click()
    # get web-element for flipkart search bar
    flpkrt_srch_wbe = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
    # Enter device to be searched in search bar
    flpkrt_srch_wbe.send_keys(str(device_type),Keys.ENTER)
```

```python
def flipkart_rev(list1,list2,list3,list4,list5,):
    # get webelements for search results
    products_wbe = driver.find_elements_by_xpath("//div[@class='_1AtVbE col-12-12']/div[@class='_13oc-S']")
    
    #iterate over each product
    for product in products_wbe:
        time.sleep(30)
#         try:
#             WebDriverWait(driver,30).until(EC.element_to_be_clickable((By.XPATH,"NEXT"))).click()
#         except :
#             WebDriverWait(driver,30).until(EC.element_to_be_clickable((By.LINK_TEXT,"NEXT"))).click()
        #click on product
        product.click()
        time.sleep(30)
        # Swtich to new wnodw
        driver.switch_to.window(driver.window_handles[1])
        try:
            # click on see all reviews button
            driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']").click()
            
            time.sleep(10)
            
            # flag if only 1 review page
            try:
                driver.find_element_by_xpath("//a[@class='ge-49M _2Kfbh8']")
            except NoSuchElementException:
                one_page_flag = True
            
            # loop for all existing review pages
            if driver.find_element_by_xpath("//a[@class='ge-49M _2Kfbh8']").text == '1' or one_page_flag == True: # check if it's 1st review page            
                # get all review tile handles
                review_tiles_wbe = driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']")
                time.sleep(10)
                # iterate over each review tile
                for tile in review_tiles_wbe:
                    # find rating and review from each tile and append to dataframe
                    if float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <= 1.0:
                        if len(list1) == 10:
                            continue
                        else:
                            list1.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                    elif float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 1.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=2.0:
                        if len(list2) == 10:
                            continue
                        else:
                            list2.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                    elif  float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 2.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=3.0:
                        if len(list3) == 10:
                            continue
                        else:
                            list3.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                    elif  float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 3.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=4.0:
                        if len(list4) == 10:
                            continue
                        else:
                            list4.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                    elif  float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 4.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=5.0:
                        if len(list5) == 10:
                            continue
                        else:
                            list5.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
            one_page_flag = False
            
            try:
                WebDriverWait(driver,30).until(EC.element_to_be_clickable((By.LINK_TEXT,"NEXT"))).click()
            except :
                WebDriverWait(driver,30).until(EC.element_to_be_clickable((By.LINK_TEXT,"NEXT"))).click()
            
            time.sleep(20)

            # get range to iterate for available pages
            for i in range(len(driver.find_elements_by_xpath("//a[contains(@class,'ge-49M')]")) - 1 ):
                time.sleep(3)
                # get all review tile handles
                review_tiles_wbe_2 = []
                review_tiles_wbe_2.clear()
                review_tiles_wbe_2 = driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']")
                
                time.sleep(5)
                try:
                    # iterate over each review tile
                    for tile in review_tiles_wbe_2:
                        # find rating and review from each tile and append to dataframe
                        if float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <= 1.0:
                            if len(list1) == 10:
                                continue
                            else:
                                list1.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                        elif float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 1.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=2.0:
                            if len(list2) == 10:
                                continue
                            else:
                                list2.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                        elif  float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 2.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=3.0:
                            if len(list3) == 10:
                                continue
                            else:
                                list3.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                        elif  float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 3.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=4.0:
                            if len(list4) == 10:
                                continue
                            else:
                                list4.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                        elif  float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) > 4.0 and float(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text) <=5.0:
                            if len(list5) == 10:
                                continue
                            else:
                                list5.append(tile.find_element_by_xpath(".//div[@class='_3LWZlK _1BLPMq']").text+"."+tile.find_element_by_xpath(".//p[@class='_2-N8zT']").text + tile.find_element_by_xpath(".//div[@class='t-ZTKy']").text)
                    time.sleep(30)
                    # click on next page button
                    try:
                        WebDriverWait(driver,30).until(EC.element_to_be_clickable((By.LINK_TEXT,"NEXT"))).click()
                    except :
                        WebDriverWait(driver,30).until(EC.element_to_be_clickable((By.LINK_TEXT,"NEXT"))).click()
                except:
                    continue
                
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
        except:
            # close current tab and switch to search page
            driver.close()
            driver.switch_to.window(driver.window_handles[0])

    return list1,list2,list3,list4,list5
```

```python
def get_reviews_snapdeal(x,list1,list2,list3,list4,list5):
    """ function to get reviews from snapdeal pages
        x - flag to check if multi-page or single-page reviews"""
    if x == 1 : # condition for multi-page reviews
        # check last page number
        page_num_wbes = WebDriverWait(driver,20).until(EC.presence_of_element_located((By.XPATH,"//div[@class='pagination']"))).find_elements_by_xpath(".//ul/li")
        # get first and last page number
        first_page_num = int(page_num_wbes[1].text)
        last_page_num = int(page_num_wbes[-2].text)
        # click on first review page
        driver.implicitly_wait(10)
        page_num_wbes[1].click()
        # iterate for each page
        for i in range(last_page_num):
            # initiate rating and reviews lists for each loop and at the end of loop append them to a final list
            ratings = []
            ratings.clear()
            reviews = []
            reviews.clear()
            
            rating_wbes = []
            review_wbes_1 = []
            review_wbes_2 = []
            
            # get rating stars webelements
            rating_wbes = WebDriverWait(driver,30).until(EC.presence_of_element_located((By.XPATH,"//div[@class='commentreview']"))).find_elements_by_xpath(".//div[@class='rating']")
            # get length of rating stars as ratings
            for rating_wbe in rating_wbes :
#                 stars_wbes = rating_wbe.find_elements_by_xpath(".//i")
                ratings.append(len(rating_wbe.find_elements_by_xpath(".//i")))
                
            # get reviews
            review_wbes_1 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='commentreview']"))).find_elements_by_xpath(".//div[@class='head']")
            review_wbes_2 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='commentreview']"))).find_elements_by_xpath(".//div[@class='user-review']/p")
            # loop
            for i in range(len(review_wbes_1)):
                reviews.append(review_wbes_1[i].text+" . "+review_wbes_2[i].text)
#             pdb.set_trace()
            # append reviews as per ratings
            for i in range(len(ratings)):
                if float(ratings[i]) <= 1.0:
                    if len(list1) == 10:
                        continue
                    else:
                        list1.append(str(ratings[i])+"."+reviews[i])
                elif float(ratings[i]) > 1.0 and float(ratings[i]) <=2.0:
                    if len(list2) == 10:
                        continue
                    else:
                        list2.append(str(ratings[i])+"."+reviews[i])
                elif  float(ratings[i]) > 2.0 and float(ratings[i]) <=3.0:
                    if len(list3) == 10:
                        continue
                    else:
                        list3.append(str(ratings[i])+"."+reviews[i])
                elif  float(ratings[i]) > 3.0 and float(ratings[i]) <=4.0:
                    if len(list4) == 10:
                        continue
                    else:
                        list4.append(str(ratings[i])+"."+reviews[i])
                elif  float(ratings[i]) > 4.0 and float(ratings[i]) <=5.0:
                    if len(list5) == 10:
                        continue
                    else:
                        list5.append(str(ratings[i])+"."+reviews[i])             
###########     pdb.set_trace()
            try:
                WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='pagination']"))).find_element_by_xpath(".//ul/li[@class='last']/a").click() # it's not working
            except Exception as e:
                print(e)
    elif x == 2: # one page reviews product
        ratings = []
        ratings.clear()
        reviews = []
        reviews.clear()

        rating_wbes = []
        review_wbes_1 = []
        review_wbes_2 = []
        # get rating stars webelements
        rating_wbes = WebDriverWait(driver,30).until(EC.presence_of_element_located((By.XPATH,"//div[@class='commentreview']"))).find_elements_by_xpath(".//div[@class='rating']")
        # get length of rating stars as ratings
        for rating_wbe in rating_wbes :
#                 stars_wbes = rating_wbe.find_elements_by_xpath(".//i")
            ratings.append(len(rating_wbe.find_elements_by_xpath(".//i")))

        # get reviews
        review_wbes_1 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='commentreview']"))).find_elements_by_xpath(".//div[@class='head']")
        review_wbes_2 = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//div[@class='commentreview']"))).find_elements_by_xpath(".//div[@class='user-review']/p")
        # loop
        for i in range(len(review_wbes_1)):
            reviews.append(review_wbes_1[i].text+" . "+review_wbes_2[i].text)
        # append reviews as per ratings
        for i in range(len(ratings)):
            if float(ratings[i]) <= 1.0:
                if len(list1) == 10:
                    continue
                else:
                    list1.append(str(ratings[i])+"."+reviews[i])
            elif float(ratings[i]) > 1.0 and float(ratings[i]) <=2.0:
                if len(list2) == 10:
                    continue
                else:
                    list2.append(str(ratings[i])+"."+reviews[i])
            elif  float(ratings[i]) > 2.0 and float(ratings[i]) <=3.0:
                if len(list3) == 10:
                    continue
                else:
                    list3.append(str(ratings[i])+"."+reviews[i])
            elif  float(ratings[i]) > 3.0 and float(ratings[i]) <=4.0:
                if len(list4) == 10:
                    continue
                else:
                    list4.append(str(ratings[i])+"."+reviews[i])
            elif  float(ratings[i]) > 4.0 and float(ratings[i]) <=5.0:
                if len(list5) == 10:
                    continue
                else:
                    list5.append(str(ratings[i])+"."+reviews[i])
    
    return list1,list2,list3,list4,list5
```
